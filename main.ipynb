```python
def analisar_eficiencia_economia(energia_consumida, energia_solar_gerada, custo_energia, preco_solar_instalacao):
    # Cálculo da eficiência energética em porcentagem
    eficiencia_energetica = (energia_solar_gerada / energia_consumida) * 100

    # Cálculo da economia anual proporcionada pela energia solar
    economia_anual = (energia_consumida - energia_solar_gerada) * custo_energia

    # Cálculo do período de retorno do investimento (payback)
    payback = preco_solar_instalacao / economia_anual if economia_anual > 0 else float('inf')

    resultado = {
        'Eficiência Energética': eficiencia_energetica,
        'Economia Anual ($)': economia_anual,
        'Período de Retorno do Investimento (Payback, anos)': payback,
    }

    # Escrever o resultado no arquivo de texto
    with open('analise_energia_solar.txt', 'w') as arquivo:
        arquivo.write("Análise de Eficiência Energética e Economia\n")
        arquivo.write(50 * "-" + "\n")
        for chave, valor in resultado.items():
            arquivo.write(f"{chave}: {valor}\n")

# Exemplo de uso da função
energia_consumida = 10000  # kWh
energia_solar_gerada = 5000  # kWh
custo_energia = 0.12  # custo por kWh em dólares
preço_solar_instalação = 15000  # custo da instalação em dólares

analisar_eficiencia_economia(energia_consumida, energia_solar_gerada, custo_energia, preço_solar_instalação)

print("Análise salva no arquivo 'analise_energia_solar.txt'")
```
